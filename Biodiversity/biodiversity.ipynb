```python
data1=pd.read_csv("PNAS_Wieczynski_2018_MainPlotMetadata.csv")
data1
gdf1 = gpd.GeoDataFrame(data1, geometry=gpd.points_from_xy(data1.Longitude, data1.Latitude))
gdf1
```

```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
# 假设gdf和gdf1是GeoDataFrame，并已经正确读取数据。

# 创建画布和轴对象，设置投影为 PlateCarree（适合全球地图的投影）
fig, ax = plt.subplots(figsize=(15, 10), subplot_kw={'projection': ccrs.PlateCarree()})

# 绘制世界地图
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
world.plot(ax=ax, color='white')

# 设置显示范围，去除南极
ax.set_extent([-180, 180, -60, 90], crs=ccrs.PlateCarree())

# 绘制gdf数据，颜色根据'S'列的值映射
gdf.plot(ax=ax, column='S', color='green', markersize=0.1, legend=True,
         legend_kwds={'label': "S Value", 'orientation': "horizontal"})

# 绘制gdf1数据，颜色根据'Species_Richness'列的值映射
gdf1.plot(ax=ax, column='Species_Richness', color='#8B0000', markersize=3)

# 添加国界线和海岸线
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.COASTLINE)

# 添加经纬度网格线
gl = ax.gridlines(draw_labels=True,color='none')
gl.top_labels = False  # 不显示顶部的标签
gl.right_labels = False  # 不显示右侧的标签
gl.xformatter = LONGITUDE_FORMATTER  # 设置经度的格式
gl.yformatter = LATITUDE_FORMATTER  # 设置纬度的格式
gl.xlabel_style = {'size': 12, 'color': 'black'}  # 设置纬度标签的风格
gl.ylabel_style = {'size': 12, 'color': 'black'}  # 设置经度标签的风格
# 显示图表
plt.show()
```
